```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
a=6
b=0.4
def dydt(x,t):
  return k(x)*x
def k(x):
  return a/x+b
t = np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline 
plt.plot(t, solve, label = r"x'=k(x)*x")
plt.grid(True)
plt.legend()
plt.show()
```
