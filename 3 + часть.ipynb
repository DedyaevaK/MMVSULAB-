```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import odeint
url = 'https://raw.githubusercontent.com/DedyaevaK/MMVSULAB1/main/Kazan.csv'
df = pd.read_csv(url)
print(df)
Years = np.array(df['Years'])
Population = np.array(df['Population'])
def dydt(Years,Population):
  k = 0.4
  return k*Years

def dydt(x,t):
  k = 0.00094
  return k*x
t = np.array(df['Years'])
solve = odeint(dydt, 7000, t=t)
solve = solve.flatten()
%matplotlib inline 

plt.plot(t, solve, label = r"x′=kx")

plt.plot(Years, Population)
plt.grid(True)
plt.title('Численность населения г.Казань', fontsize=13)
```
